```python
import numpy as np
from numpy import load
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## K means clusturing -> dev_stream_sliced_w_SAD

```python
d_vector_dev_str_sad = load('d_vector_dev_seg_fbank.npy')
X_sad = np.array(d_vector_dev_str_sad)
kmeans = KMeans(n_clusters=167, random_state=0).fit(X_sad)
kmeans.labels_

c_dev_sad = kmeans.cluster_centers_
cl_dev_sad = kmeans.labels_
labels_dev_sad = kmeans.labels_

for i in range(len(labels_dev_sad)):
    print(labels_dev_sad[i])
```

```python
counter = []
for i in labels_dev_sad:
    if i == 164:
        counter.append(i)
len(counter)
```

```python
unique, counts = np.unique(cl_dev_sad, return_counts=True)
dict(zip(unique, counts))
```

### tSNE

```python
X_embedded = TSNE(n_components=2).fit_transform(c_dev_sad)
df_subset = pd.DataFrame(X_embedded)
```

```python
print(__doc__)

from sklearn.cluster import kmeans_plusplus
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import random

# Generate sample data
n_samples = 4000
n_components = 166

X, y_true = make_blobs(n_samples=n_samples,
                       centers=n_components,
                       cluster_std=0.60,
                       random_state=None)
#X = X_embedded_dev
#y_true = cl_dev.astype(int)
X = X[:, ::-1]
# Calculate seeds from kmeans++
centers_init, indices = kmeans_plusplus(X, n_clusters=166, random_state=None)

# Plot init seeds along side sample data
plt.figure(1)
get_colors = lambda n: list(map(lambda i: "#" + "%06x" % random.randint(0, 0xFFFFFF),range(n)))
colors = get_colors(166)
for k, col in enumerate(colors):
    cluster_data = y_true == k
    print(cluster_data)
    plt.scatter(X[cluster_data, 0], X[cluster_data, 1], c=col, marker='.', s=10)

plt.scatter(centers_init[:, 0], centers_init[:, 1], c='b', s=50)
plt.title("K-Means++ Initialization")
plt.xticks([])
plt.yticks([])
plt.show()
```
